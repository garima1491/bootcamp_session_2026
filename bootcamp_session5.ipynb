```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
df=df.rename(columns={
    "lifeExp" : "life_expectancy",
    "pop" : "population",
    "gdpPercap" : "gdp_per_capita"
})

print(df.head())
print(df.columns)
```

```python
print("Missing values in each column:")
print(df.isna().sum())
print("Number of duplicate rows:",df.duplicated().sum())
```

```python
dirty_demo = df.head(12).copy()
#deliberately introduce missing values for teaching practice

dirty_demo.loc[2,"life_expectancy"]=np.nan
dirty_demo.loc[5,"gdp_per_capita"]=np.nan

print(dirty_demo)
print(dirty_demo.isna().sum())

#fill mising numerical values using median

clean_demo = dirty_demo.copy()
clean_demo["life_expectancy"]=clean_demo["life_expectancy"].fillna(clean_demo["life_expectancy"].median())
clean_demo["gdp_per_capita"]=clean_demo["gdp_per_capita"].fillna(clean_demo["gdp_per_capita"].median())


print(clean_demo.isna().sum())
print(clean_demo)
```

```python
#using pandas-- i want all of you to create a basic data analysis :-  # you can take use
#of llm
# year range
#number of countries
# number of continents



print("year range:",df["year"].min(),"to",df["year"].max())
print("Number of countries:",df["country"].nunique())
print("Number of continents:",df["continent"].nunique())


print("Average life expectancy:",df["life_expectancy"].mean())
print("Minimum life expectancy:",df["life_expectancy"].min())
print("Maximum life expectancy:",df["life_expectancy"].max())

print(df.describe())
```

```python
#group and count-gdp population life expectancy


print(df["continent"].value_counts())

continent_summary = df.groupby("continent")[[      "life_expectancy",
    "population",
    "gdp_per_capita"
]].mean()

print(continent_summary)
```

```python
latest_year = df["year"].max()
latest_df =df[df["year"]==latest_year].copy()

print("Latest year:",latest_year)
print(latest_df.head())
print(latest_df.shape)
```

```python
# using the lib matplotlib #I need a bar chart that show me avg life expectancy  by continent

continent_life = latest_df.groupby("continent")["life_expectancy"].mean()

plt.bar(continent_life.index,continent_life.values)
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.title("Average Life Expectancy by Continent")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df=pd.read_csv(url)

df=df.rename(columns={
    "lifeExp" : "life_expectancy",
    "pop" : "population",
    "gdpPercap" : "gdp_per_capita"
})

# temporal?
# line chart life expectancy over rime for 3 countries India ,China ,United states

selected_countries=["India","China","United States"]
trend_df =df[df["country"].isin(selected_countries)]

plt.figure(figsize=(9,5))


for country in selected_countries:
  country_data = trend_df[trend_df["country"]==country]
  plt.plot(country_data["year"],country_data["life_expectancy"],label=country)


plt.title("Life Expectancy Over Time for Selected Countries")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#GDP per Capita VS Life Expectancy in 2007

import pandas as pd
import matplotlib.pyplot as plt

# Gapminder dataset load
url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
df = pd.read_csv(url)

# 2007 ka data filter
data_2007 = df[df['year'] == 2007]

# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(data_2007['gdpPercap'],
            data_2007['lifeExp'],
            alpha=0.7)

# Log scale on x-axis
plt.xscale('log')

# Labels and title
plt.title('GDP per Capita vs Life Expectancy in 2007')
plt.xlabel('GDP per Capita - log scale')
plt.ylabel('Life Expectancy')

plt.grid(True)
plt.show()
```
