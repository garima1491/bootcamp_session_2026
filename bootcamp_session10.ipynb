```python
tokens=["the","animal","was","tired"]
attention_score_for_it={"the":0.1,"animal":0.6,"was":0.1,"tired":0.15}
attention_score_for_it
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.bar(attention_score_for_it.keys(),attention_score_for_it.values())
plt.title("Simplified Attention Scores for 'it' ")
plt.xlabel("Previous tokens")
plt.ylabel("Attention score")
plt.ylim(0, 1)
plt.show()
```

```python
highest_attention_token =max(attention_score_for_it,key=attention_score_for_it.get)

print("Token with highest attention:",highest_attention_token)
#
```
