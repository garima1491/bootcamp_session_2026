```python
import tensorflow as tf #mathmatical function,tensor manupulating tools
from tensorflow.keras import layers #keras is an api for tensorflow  jaha oprations kar sakte hai
from tensorflow.keras.models import Sequential


#Hyperprameters

vocab_size = 1000 # unique wors or unique tokens
embedding_dim = 16 # 16 vectors ka list (or array)

sequence_length =8

#model architecture of LSTM

lstm = Sequential(
    [
        #Layer 1: Input wording embeddings
        layers.Embedding(
            input_dim=vocab_size,
            output_dim=embedding_dim,
            input_length=sequence_length,
            name="embedding_layer"
        ),


        #Layer 2: sequence processing (RNN  base - with lstm touch)
        layers.LSTM(units=32 ,name="lstm_layer"),#it will process 32 internal memory cell info ---it processes []vectors



        #Layer 3: output binary classification
        layers.Dense(units=1,activation="sigmoid",name="output_layer"),

        ]
    ,name="Sequential_LSTM_Classifier"

)

#3. verification
lstm.summary()
```
