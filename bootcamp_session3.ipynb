WRITE A CODEE FOR PASS OR FAIL,PASS BE 40 OR ABOVE RUN THIS CODE ON A LIST OF STUDENT OF 5 STUDENT

```python
marks=[355,78,40,22,60]
for i in range(len(marks)):
  if marks[i]>=40:
    print("Student",i+1,"-",marks[i],":Pass")
  else:
     print("Student",i+1,'-',marks[i],":Fail")
```

WRITE A CODE FOR CHECK WHEATHER NUMBER IS POSITIVE ,NAGATIVE OR ZERO

```python
#list of number
numbers=[10,-5,0,25,-8]
for num in numbers:
  if num>0:
    print(num,"- Positive")
  elif num<0:
    print(num,"- Nagative")
  else:
    print(num,"- Zero")
```

STUDENT NAME AS INPUT
BASED ON THEIR MARKS YOU HAVE TO CONSIDER A PRINT STATEMENT FOR "PERFORMANCE" MARKS IS MORE THAN 75 THEN PERFORMANCE IS STRONG MARKS IS EQUAL TO OR GREATER THAN 40,PERFORMANCE SHOULD PRINT NEED TO IMPROVEMENT LESS
20 THEN PRINT FAIL ,REPEAT CLASS

ATTENDANCE 75-WARNING LOW ATEENDENCE STUDY_HOUR LESS THAN4 PRINT SHOULD COME WARNING MORE STUDY TIME

```python
students={
    "Garima":{"study_hours":5,"attendence":80,"marks":85},
    "Riya":{"study_hours":3,"attendence":70,"marks":60},
    "Siya":{"study_hours":2,"attendence":65,"marks":15},
    "Neha":{"study_hours":6,"attendence":90,"marks":78},
    "Kinjal":{"study_hours":4,"attendence":72,"marks":45},

}

name=input("enter student name:")
if name in students:
  s= students[name]
  print("marks:",s["marks"])

  if s["marks"]>75:
    print("performance: strong")
  elif s["marks"]>=40:
    print("performance :need improvement")
  elif s["marks"]<20:
    print("performance:fail,repeat class")



  if s["attendence"]<75:
    print("warning: needs low attendence")


  if s["study_hours"]<4:
    print("warning :needs more study time")

else:
  print("student not found")
```
